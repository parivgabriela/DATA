## Clean csv file

The first 7th rows contain nan values. That rows can be delete. The next script will delete the file without any modification in the original file, the script will copy the file except for the first rows.

```python
import csv

new_file = "clean_file"
original_file = path_to_original_file.format(name_file)
with open(original_file, 'r') as inp, open(path_to_original_file.format(new_file), 'w') as out:
  writer = csv.writer(out)
  index = 0
  for row in csv.reader(inp):
    if index > 6:
      writer.writerow(row)
    index += 1
```
